```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pathlib import Path

# اقرأ الملف المنظف
df = pd.read_csv("/Users/ahmadabotrab/Desktop/arabic_sentiment_classification/data/cleaned_dataset.csv")

# أنشئ مجلد cleaned إذا لم يكن موجودًا
Path("/Users/ahmadabotrab/Desktop/arabic_sentiment_classification/data/cleaned").mkdir(parents=True, exist_ok=True)

# split
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)

# حفظ الملفات
train_df.to_csv("/Users/ahmadabotrab/Desktop/arabic_sentiment_classification/data/cleaned/train.csv", index=False)
test_df.to_csv("/Users/ahmadabotrab/Desktop/arabic_sentiment_classification/data/cleaned/test.csv", index=False)

print("train shape:", train_df.shape)
print("test shape:", test_df.shape)
```
