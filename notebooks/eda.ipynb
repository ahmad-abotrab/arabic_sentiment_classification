```python
TEXT_COL = "text"
LABEL_COL = "class"
```

```python
cleaned_df = clean_dataframe(df, text_col=TEXT_COL, label_col=LABEL_COL)
cleaned_df.head()
```

```python
print("Original shape:", df.shape)
print("Cleaned shape:", cleaned_df.shape)
print(cleaned_df["label"].value_counts())
```

```python
raw_samples = df[[TEXT_COL, LABEL_COL]].dropna().head(5)

for _, row in raw_samples.iterrows():
    raw_text = str(row[TEXT_COL])
    cleaned_text = clean_text(raw_text)
    print("RAW    :", raw_text)
    print("CLEANED:", cleaned_text)
    print("-" * 80)
```
